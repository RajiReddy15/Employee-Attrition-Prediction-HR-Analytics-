# Importing data from sql workbench

```python
from sqlalchemy import create_engine
import pandas as pd
# MySQL connection credentials
username = "root"
password = "root123"
host = "localhost"
port = 3306
database = "hr(attrition)"  # your database name

# Correct SQLAlchemy engine (no backticks in URL)
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")


df=pd.read_sql('SELECT * FROM hr_data', con=engine)
df.head()
```

# Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("Attrition",axis=1)
y=df['Attrition']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

# Handiling Imbalance Data

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

log_model = LogisticRegression(max_iter=1000, class_weight='balanced')
rf_model = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced')

log_model.fit(x_train_res, y_train_res)
rf_model.fit(x_train_res, y_train_res)
```

```python
from sklearn.metrics import classification_report,  roc_auc_score

y_pred_log = log_model.predict(x_test)
y_pred_rf = rf_model.predict(x_test)

print("Logistic Regression:")
print(classification_report(y_test, y_pred_log))
print("ROC-AUC:", roc_auc_score(y_test, log_model.predict_proba(x_test)[:,1]))

print("Random Forest:")
print(classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, rf_model.predict_proba(x_test)[:,1]))
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Define model
xgb_model = XGBClassifier(
    n_estimators=300,        # number of trees
    max_depth=4,             # tree depth
    learning_rate=0.1,       # step size shrinkage
    subsample=0.8,           # subsample ratio for training instance
    colsample_bytree=0.8,    # subsample ratio for features
    scale_pos_weight=(len(y_train_res) - sum(y_train_res)) / sum(y_train_res),  
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

# Train on resampled data (or directly on imbalanced data with scale_pos_weight)
xgb_model.fit(x_train_res, y_train_res)
# Predictions
y_pred_xgb = xgb_model.predict(x_test)
y_pred_prob_xgb = xgb_model.predict_proba(x_test)[:,1]

# Evaluation
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_pred_prob_xgb))

```
