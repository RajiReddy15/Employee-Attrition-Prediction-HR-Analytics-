# Droping unnecessary columns and secondary columns

```python
# Drop unnecessary columns
df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1, inplace=True)
secondary_columns = [
    'DailyRate',
    'Education',
    'EducationField',
    'HourlyRate',
    'JobInvolvement',
    'MaritalStatus',
    'MonthlyRate',
    'PerformanceRating',
    'StockOptionLevel'
]

# Drop secondary columns
df = df.drop(columns=secondary_columns)
```

# Saving the data into new table in mysql workbench

```python
from sqlalchemy import create_engine
# MySQL connection credentials
username = "root"
password = "root123"
host = "localhost"
port = 3306
database = "hr(attrition)"  # your database name

# Correct SQLAlchemy engine (no backticks in URL)
engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")

# Insert data into MySQL table
df.to_sql('employeeattrition_main', con=engine, if_exists='replace', method='multi', index=False,chunksize=1000 )
```

```python
# retriving data from the new table to verify
df_new = pd.read_sql("SELECT * FROM employeeattrition_main", con=engine)
df_new.head()
```
